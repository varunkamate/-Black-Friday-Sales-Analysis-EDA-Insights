Basic EDA on train data set

Importing all required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("train.csv")
```

Exploring train data set

```python
df.describe()
```

```python
#Are there any missing values in each column
df.isna().sum()
```

```python
#What is the percentage of missing data in Product_Category_2 and Product_Category_3?

Product_Category_2=(df['Product_Category_2'].isna().sum()/len(df))*100
Product_Category_3=(df['Product_Category_3'].isna().sum()/len(df))*100
print(f"Percentage of missing data in Product_Category_2 is: {round(Product_Category_2,2)}%")
print(f"percentage of missing data in Product_Category_3 is: {round(Product_Category_3,2)}%")
```

```python
#Are there any duplicate rows in the dataset?

df.duplicated().sum() #so no duplicate values in the data set
```

```python
#What is the data type of each column, and are they appropriate?
df.dtypes
```

```python
#Are there any outliers in Purchase values?
plt.figure(figsize=(6,4))
sns.boxplot(df["Purchase"])
plt.title("Outeliers for purchase column")
plt.show() #so  their are some outliers in the "Purchase" column
```

```python
#What is the gender distribution of customers?
plt.figure(figsize=(6,3))
gender=df["Gender"].value_counts()
sns.barplot(x=gender.index,y=gender.values)
plt.title("Gender distribution of customers")
plt.xlabel("Gender")
plt.ylabel("Number of customers")
plt.show()
```

OR

```python
sns.countplot(x="Gender", data=df,hue="Gender")
plt.title("Gender distribution of customers")
plt.xlabel("Gender")
plt.ylabel("Number of customers")
plt.show()
```

```python
##Which age group has the highest number of purchases?
sns.countplot(data=df, x="Age", order=df["Age"].value_counts().index,hue="Age")
plt.title("Number of Purchases by Age Group")
plt.show()
```

```python
#How is purchase amount distributed across different Occupations?
sns.barplot(data=df,x="Occupation",y="Purchase",hue="Occupation")
plt.title("Purchase amount distributed across different Occupations")
plt.xlabel("Different Occupations")
plt.ylabel("Purchase Amount")
plt.show()
```

```python
#Which City_Category has the maximum customers?
max_cust_by_city_cat=df.groupby("City_Category")["User_ID"].nunique()
print(f"City_Category has the maximum customers is :- {max_cust_by_city_cat.idxmax()}:{max_cust_by_city_cat.max()}")
```

```python
#Does Stay_In_Current_City_Years impact the purchase amount?
sns.barplot(data=df,x="Stay_In_Current_City_Years",y="Purchase",estimator="mean",hue="Stay_In_Current_City_Years")
plt.show() #No impact of it
```

```python
#Which Product_Category_1 is most frequently purchased?
most_pur=df["Product_Category_1"].value_counts()
print(f"From Product_Category_1 most frequently purchased is {most_pur.idxmax()} with count {most_pur.max()}")
```

```python
#Which product (Product_ID) generated the highest total sales?
high_sale=df.groupby("Product_ID")["Purchase"].sum()
print(f"Product (Product_ID) generated the highest total sales is : {high_sale.idxmax()} with total sale amount: {high_sale.max()}")
```

```python
#Which products are bought most often by different age groups?
most_bought=df.groupby("Product_ID")["Age"].nunique()
print(most_bought.head())
```

```python
#What is the distribution of Purchase amount for different product categories?
plt.figure(figsize=(12,6))
plt.subplot(2,2,1)
plt.suptitle("Purchase Amount Distribution by Product_Category_1,Product_Category_2,Product_Category_3")
sns.boxplot(data=df, x="Product_Category_1", y="Purchase")

plt.subplot(2,2,2)
sns.boxplot(data=df, x="Product_Category_2", y="Purchase")

plt.subplot(2,2,3)
sns.boxplot(data=df, x="Product_Category_3", y="Purchase")
plt.show()

```

```python
#What is the average purchase amount by gender?
avg_pur=df.groupby("Gender")["Purchase"].mean()
print(f"Average purchase amount by gender is : {round(avg_pur,2)}")
```

```python
#Does marital status influence the purchase amount?
sns.boxplot(data=df,x="Marital_Status",y="Purchase")
plt.title("Box plot for marital status influence the purchase amount")
plt.show() #So as shown in the plot below it does not influence the purchase amount
```

```python
#Which age group spends the most money on average?
avg_pur_grp=df.groupby("Age")["Purchase"].mean()
print(avg_pur_grp)
```

```python
#Do customers in metropolitan cities spend more than those in smaller cities?

city_avg_pur=df.groupby("City_Category")["Purchase"].mean()
print(city_avg_pur)
```

```python
#How many unique products does a single user purchase on average?

unique_product=df.groupby("User_ID")["Product_ID"].nunique()
single_user_pur=unique_product.mean()
print(f"Unique products does a single user purchase on average is: {round(single_user_pur,2)}")
```

```python
pd.crosstab(df["Age"], df["Product_Category_1"])
```

```python
#Are Occupation and City_Category related to purchase behavior?
m=["*","o",">"]
sns.lineplot(data=df,x="Occupation",y="Purchase",estimator="mean",hue="City_Category",markers=m)
plt.title("Occupation and City_Category related to purchase behavior line plot")
plt.show() #so the occupation and city_category related to the purchase behaviour
```

```python
#Is there any relationship between Stay_In_Current_City_Years and product categories purchased?

pd.crosstab(df["Stay_In_Current_City_Years"], df["Product_Category_1"])
```

```python
#Compare average purchase amounts across gender â€” which gender tends to spend more?
plt.figure(figsize=(10,6))
sns.barplot(data=df,x="Gender",y="Purchase",estimator="mean",hue="Gender")
plt.title("Bar plot for to compare average purchase amounts across gender")
plt.xlabel("Gender")
plt.ylabel("Average purchase amount")
plt.show() #so male spend more 
```

```python
#See which occupations spend the most and least on average.
avg_amt_ocu=df.groupby("Occupation")["Purchase"].mean()
print(f"Occupations spend the most average is {avg_amt_ocu.idxmax()} with average of {round(avg_amt_ocu.max(),2)}")
print(f"Occupations spend the least average is {avg_amt_ocu.idxmin()} with average of {round(avg_amt_ocu.min(),2)}")
```

```python
#Use correlation heatmaps to check relationships between numeric features and Purchase.

h=df[['Occupation','Marital_Status', 'Product_Category_1',
       'Product_Category_2', 'Product_Category_3','Purchase']]
sns.heatmap(h)
plt.show()
```

```python
#See if occupation influences purchase behavior differently across city categories.
plt.figure(figsize=(10,6))
sns.barplot(data=df,x="Occupation",y="Purchase",estimator="mean",hue="City_Category")
plt.title("bar plot for occupation influences purchase behavior differently across city categories")
plt.show()
```

```python
#Identify the top-spending gender within each city category
plt.figure(figsize=(10,6))
sns.barplot(data=df,x="City_Category",y="Purchase",estimator="mean",hue="Gender")
plt.title("Bar plot for to identify the top-spending gender within each city category")
plt.show()
```

```python
#Find the most popular product categories for each age group.
most_frq=df.groupby("Age")["Product_Category_1"].apply(lambda x:x.mode()[0])
print(most_frq)
```
